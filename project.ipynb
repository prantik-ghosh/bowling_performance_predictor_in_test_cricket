#### Concatenating dataframes

```python
frames = [df111,df112,df121,df122,df211,df212,df221,df222]
df = pd.concat(frames)
df.shape
```

```python
df.head()
```

```python
print(df['Home_Away'].value_counts())
print(df['Bowling_Arm'].value_counts())
print(df['Pace_Spin'].value_counts())
```

#### Cleaning "Opposition" column

```python
df['Opposition'] = df.Opposition.apply(lambda x: x[2:])
df.head()
```

#### Manufacturing "PlayerName", "Country", and "Year" columns

```python
df['PlayerName'] = df.Player.apply(lambda x: x[:x.rfind('(')-1])
df.head()
```

```python
df['Country'] = df.Player.apply(lambda x: x[x.rfind('(')+1:x.rfind(')')])
df.head()
```

```python
df['Year'] = df['Start Date'].apply(lambda x: int(x[-4:]))
df.head()
```

#### Dropping redundant columns "Player" and "Start Date"

```python
df.drop('Player',axis=1,inplace=True)
df.drop('Start Date',axis=1,inplace=True)
df.head()
```

```python
df.shape
```

#### Dropping records where "Country" or "Opposition" is not among the major 7 countries

```python
df.Country.value_counts()
```

```python
df.Opposition.value_counts()
```

```python
mask = df['Country'].apply(lambda x: x in ['ENG','AUS','WI','SA','INDIA','SL','NZ'])
df = df[mask]
df.shape
```

```python
mask2 = df['Opposition'].apply(lambda x: x in ['England','Australia','India','South Africa',
                                              'West Indies','Sri Lanka','New Zealand'])
df = df[mask2]
df.shape
```

#### Mapping the country codes in the "Country" column to the country names

```python
country_dict = {'ENG':'England', 'AUS':'Australia', 'INDIA':'India', 'SA':'South Africa',
               'WI':'West Indies', 'NZ':'New Zealand', 'SL':'Sri Lanka'}
country_dict
```

```python
df['Country'].replace(country_dict, inplace=True)
df.head()
```
